```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('Parallel_time.dat')
cores     = data[:,0] # column 0 are the values of N
time = data[:,1] # column 1 are the values of error

plt.plot(cores, time, color = "blue")
plt.scatter(cores, time, color = "red")
plt.xlabel("Cores")
plt.ylabel("Time(secs)")
plt.title("Strong Scaling")
plt.savefig('scaling.png')
plt.show()
```

```python
plt.plot(cores, time[0]/time, color = "blue")
plt.scatter(cores, time[0]/time, color = "red")
plt.xlabel("Cores")
plt.ylabel("Speedup")
plt.title("Speedup Graph")
plt.savefig('speedup.png')
plt.show()
```

```python
x = cores
y1 = time
y2 = time[0]/time
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, color='green')
ax1.scatter(x, y1, color='red')
ax2.plot(x, y2, color='blue')
ax2.scatter(x, y2, color='black')

ax1.set_xlabel('Cores')
ax1.set_ylabel('Time(secs)')
ax2.set_ylabel('Speedup')
plt.title('Scaling and Speedup')
plt.savefig('Dual_Axis.png')
plt.show()
```
